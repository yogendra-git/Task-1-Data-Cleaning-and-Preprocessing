```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
```

```python
print(df)
```

```python
missing_value = df.isnull()
```

```python
print(missing_value)
```

```python
df_fillna = df.fillna(0)
```

```python
print(df_fillna)
```

```python
duplicate_row_drop = df.drop_duplicates()
```

```python


df['type'] = df['type'].str.strip().str.title()


df['country'] = df['country'].fillna('Unknown').str.strip().str.title()


country_replacements = {
    'United States': 'USA',
    'United States Of America': 'USA',
    'United Kingdom': 'UK',
    'South Korea': 'South Korea',
    'Russia': 'Russian Federation',
    'Hong Kong S.A.R.': 'Hong Kong',
    'Usa': 'USA',
    'Uk': 'UK',
    'India ': 'India',  
}

df['country'] = df['country'].replace(country_replacements)


df['rating'] = df['rating'].fillna('Unrated').str.strip().str.upper()

rating_replacements = {
    'UR': 'UNRATED',
    'NR': 'UNRATED',
    'TVMA': 'TV-MA',
    'TV MA': 'TV-MA',
    'TVMA ': 'TV-MA',
    'PG13': 'PG-13',
    'PG 13': 'PG-13',
    'TVPG': 'TV-PG',
    'TV Y': 'TV-Y',
    'TVY': 'TV-Y',
    'TVG': 'TV-G'
}

df['rating'] = df['rating'].replace(rating_replacements)


df['listed_in'] = df['listed_in'].fillna('Unknown').str.strip().str.title()


df['genres'] = df['listed_in'].str.split(', ')


df['director'] = df['director'].fillna('Unknown').str.strip().str.title()
df['cast'] = df['cast'].fillna('Unknown').str.strip().str.title()


df['cast_list'] = df['cast'].str.split(', ')


print(df[['type', 'country', 'rating', 'listed_in', 'director', 'cast']].sample(5))
```

```python
 from datetime import datetime
 df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
df['date_added'] = df['date_added'].fillna('Unknown')
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python

print("Before Fixing:\n", df.dtypes)


df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')  # nullable integer


df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')


df[['duration_num', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_num'] = pd.to_numeric(df['duration_num'], errors='coerce')


print("\nAfter Fixing:\n", df.dtypes)
```
